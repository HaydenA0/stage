```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import plotly.express as px

# Set plot style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6) # Default figure size

# Load the dataset
data_raw = pd.read_csv("flights.csv")

# --- Cleaning and Basic Feature Engineering (from previous discussion) ---
columns_to_drop = [
    'FLT_DEP_IFR_2', 'FLT_ARR_IFR_2', 'FLT_TOT_IFR_2',
    'MONTH_MON', 'Pivot Label', 'APT_NAME'
]
data_cleaned = data_raw.drop(columns=columns_to_drop)

data_cleaned['FLT_DATE'] = pd.to_datetime(data_cleaned['FLT_DATE'])
data_cleaned['day_of_week'] = data_cleaned['FLT_DATE'].dt.dayofweek # Monday=0, Sunday=6
data_cleaned['day_of_month'] = data_cleaned['FLT_DATE'].dt.day
data_cleaned['day_of_year'] = data_cleaned['FLT_DATE'].dt.dayofyear
data_cleaned['week_of_year'] = data_cleaned['FLT_DATE'].dt.isocalendar().week.astype(int)
data_cleaned['quarter'] = data_cleaned['FLT_DATE'].dt.quarter
data_cleaned['is_weekend'] = (data_cleaned['FLT_DATE'].dt.dayofweek >= 5).astype(int)

# --- NEW FEATURE ENGINEERING FOR PANDEMIC CONTEXT AND AIRPORT BASELINES ---

# 1. Create a 'pandemic_phase' feature
# Based on the Eurostat plots and general knowledge of the pandemic timeline
def assign_pandemic_phase(row):
    year = row['YEAR']
    month = row['MONTH_NUM']
    if year < 2020:
        return 'Pre-Pandemic (2016-2019)'
    elif year == 2020:
        if month >= 3 and month <= 5: # March-May 2020: Initial severe drop
            return 'Pandemic Peak Drop (Spring 2020)'
        elif month >= 6 and month <= 8: # June-Aug 2020: Slight summer recovery
            return 'Pandemic Initial Recovery (Summer 2020)'
        else: # Sep-Dec 2020: Second wave / Fall-Winter stagnation
            return 'Pandemic Second Wave (Fall/Winter 2020)'
    elif year == 2021:
        if month >= 1 and month <= 4: # Jan-April 2021: Continued stagnation
            return 'Pandemic Continued Stagnation (Early 2021)'
        elif month >= 5 and month <= 8: # May-Aug 2021: Stronger recovery
            return 'Pandemic Strong Recovery (Summer 2021)'
        else: # Sep-Dec 2021: Post-summer recovery continues
            return 'Pandemic Post-Summer Recovery (Late 2021)'
    elif year == 2022: # Up to May 2022 in dataset
        return 'Post-Pandemic Recovery (2022)'
    return 'Other' # Should not happen with this data

data_cleaned['pandemic_phase'] = data_cleaned.apply(assign_pandemic_phase, axis=1)

# Ensure the order of pandemic phases for better visualization
pandemic_phase_order = [
    'Pre-Pandemic (2016-2019)',
    'Pandemic Peak Drop (Spring 2020)',
    'Pandemic Initial Recovery (Summer 2020)',
    'Pandemic Second Wave (Fall/Winter 2020)',
    'Pandemic Continued Stagnation (Early 2021)',
    'Pandemic Strong Recovery (Summer 2021)',
    'Pandemic Post-Summer Recovery (Late 2021)',
    'Post-Pandemic Recovery (2022)'
]
data_cleaned['pandemic_phase'] = pd.Categorical(data_cleaned['pandemic_phase'], categories=pandemic_phase_order, ordered=True)

# 2. Calculate average pre-pandemic traffic for each airport and month
# This feature serves as a baseline for 'normal' traffic for each airport/month combination
pre_pandemic_data = data_cleaned[data_cleaned['YEAR'] < 2020]
avg_traffic_pre_pandemic_monthly = pre_pandemic_data.groupby(['APT_ICAO', 'MONTH_NUM'])['FLT_TOT_1'].mean().reset_index()
avg_traffic_pre_pandemic_monthly.rename(columns={'FLT_TOT_1': 'avg_pre_pandemic_monthly_traffic'}, inplace=True)

# Merge this back into the main DataFrame
data_cleaned = pd.merge(data_cleaned, avg_traffic_pre_pandemic_monthly,
                        on=['APT_ICAO', 'MONTH_NUM'], how='left')

# Handle cases where an airport might not have pre-pandemic data (rare, but possible)
# For simplicity, fill NaNs with the overall average pre-pandemic traffic or 0
data_cleaned['avg_pre_pandemic_monthly_traffic'].fillna(
    pre_pandemic_data['FLT_TOT_1'].mean(), inplace=True
)

# 3. Create 'traffic_pct_of_pre_pandemic' (your "mixing columns" idea)
# This captures the *relative* recovery/decline, useful for normalizing across different airport sizes
data_cleaned['traffic_pct_of_pre_pandemic'] = (
    data_cleaned['FLT_TOT_1'] / data_cleaned['avg_pre_pandemic_monthly_traffic']
) * 100

# Handle potential division by zero if avg_pre_pandemic_monthly_traffic was 0
data_cleaned['traffic_pct_of_pre_pandemic'].replace([np.inf, -np.inf], np.nan, inplace=True)
data_cleaned['traffic_pct_of_pre_pandemic'].fillna(0, inplace=True) # If no pre-pandemic traffic, then current traffic is 0% of that

# 4. Create 'airport_volume_category' based on overall pre-pandemic average
# First, calculate overall pre-pandemic average traffic for each airport
avg_traffic_pre_pandemic_overall = pre_pandemic_data.groupby('APT_ICAO')['FLT_TOT_1'].mean().reset_index()
avg_traffic_pre_pandemic_overall.rename(columns={'FLT_TOT_1': 'avg_traffic_pre_pandemic_overall'}, inplace=True)

data_cleaned = pd.merge(data_cleaned, avg_traffic_pre_pandemic_overall,
                        on='APT_ICAO', how='left')

data_cleaned['avg_traffic_pre_pandemic_overall'].fillna(
    pre_pandemic_data['FLT_TOT_1'].mean(), inplace=True
)

# Create airport volume categories based on quartiles of avg_traffic_pre_pandemic_overall
data_cleaned['airport_volume_category'] = pd.qcut(
    data_cleaned['avg_traffic_pre_pandemic_overall'],
    q=4, # 4 categories: Q1, Q2, Q3, Q4
    labels=['Small', 'Medium', 'Large', 'Very Large'],
    duplicates='drop' # Handle cases where there aren't enough unique values for 4 quartiles
)
# Ensure order for plotting
airport_volume_order = ['Small', 'Medium', 'Large', 'Very Large']
data_cleaned['airport_volume_category'] = pd.Categorical(data_cleaned['airport_volume_category'], categories=airport_volume_order, ordered=True)

# --- Drop original FLT_DATE column now that features are extracted ---
data_cleaned = data_cleaned.drop(columns=['FLT_DATE'])

print("Data Cleaning and Feature Engineering Complete.")
print("New columns available:", [col for col in data_cleaned.columns if col not in data_raw.columns])

# --- Re-run Visualizations ---

# 1. Traffic by Pandemic Phase
plt.figure(figsize=(14, 7))
sns.boxplot(x='pandemic_phase', y='FLT_TOT_1', data=data_cleaned)
plt.title('Total Flights by Pandemic Phase')
plt.xlabel('Pandemic Phase')
plt.ylabel('Number of Flights (FLT_TOT_1)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("figure4.png", bbox_inches='tight')

plt.show()

# Traffic as Percentage of Pre-Pandemic Average by Phase (a "mixed column" feature)
plt.figure(figsize=(14, 7))
sns.boxplot(x='pandemic_phase', y='traffic_pct_of_pre_pandemic', data=data_cleaned)
plt.title('Traffic as Percentage of Pre-Pandemic Monthly Average by Phase')
plt.xlabel('Pandemic Phase')
plt.ylabel('Traffic (% of Pre-Pandemic Monthly Average)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("figure5.png", bbox_inches='tight')

plt.show()

# 2. Total Flights by Airport Volume Category
plt.figure(figsize=(12, 7))
sns.boxplot(x='airport_volume_category', y='FLT_TOT_1', data=data_cleaned)
plt.title('Total Flights by Airport Volume Category')
plt.xlabel('Airport Volume Category (based on Pre-Pandemic Average)')
plt.ylabel('Number of Flights (FLT_TOT_1)')
plt.savefig("figure6.png", bbox_inches='tight')

plt.show()

# 3. Visualizing Interactions: Day of Week vs. Airport Volume Category
plt.figure(figsize=(14, 7))
sns.boxplot(x='day_of_week', y='FLT_TOT_1', hue='airport_volume_category',
            data=data_cleaned, palette='viridis') # Removed `order` for hue, as it's set in categorical type.
plt.title('Total Flights by Day of Week, Grouped by Airport Volume Category')
plt.xlabel('Day of Week (0=Mon, 6=Sun)')
plt.ylabel('Number of Flights (FLT_TOT_1)')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.legend(title='Airport Volume')
plt.tight_layout()
plt.savefig("figure7.png", bbox_inches='tight')

plt.show()

# 4. Interaction Heatmap: Average Traffic by Year and Month
avg_traffic_ym = data_cleaned.groupby(['YEAR', 'MONTH_NUM'])['FLT_TOT_1'].mean().unstack()

plt.figure(figsize=(12, 8))
sns.heatmap(avg_traffic_ym, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=.5)
plt.title('Average Total Flights by Year and Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.savefig("figure8.png", bbox_inches='tight')

plt.show()

# --- Corrected PCA Plotting Code (from previous answer) ---
numerical_features_for_pca = [
    'YEAR', 'MONTH_NUM', 'day_of_week', 'day_of_month', 'day_of_year',
    'week_of_year', 'quarter', 'is_weekend',
    # Include new numerical features for PCA if desired, e.g., 'avg_pre_pandemic_monthly_traffic',
    # 'traffic_pct_of_pre_pandemic', 'avg_traffic_pre_pandemic_overall'
    # For now, let's stick to the original set to analyze date effects only.
]

X_pca_raw = data_cleaned[numerical_features_for_pca]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_pca_raw)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(data = principal_components,
                      columns = ['principal_component_1', 'principal_component_2'])

pca_df['FLT_TOT_1'] = data_cleaned['FLT_TOT_1'].values
pca_df['APT_ICAO'] = data_cleaned['APT_ICAO'].values

plt.figure(figsize=(12, 8))
ax = sns.scatterplot(x='principal_component_1', y='principal_component_2',
                     hue='FLT_TOT_1', palette='viridis',
                     size='FLT_TOT_1', sizes=(10, 400),
                     alpha=0.6, data=pca_df, legend=False)

plt.title('PCA of Numerical Flight Features (Colored by FLT_TOT_1)')
plt.xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]*100:.2f}% variance)')
plt.ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]*100:.2f}% variance)')

norm = plt.Normalize(pca_df['FLT_TOT_1'].min(), pca_df['FLT_TOT_1'].max())
sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
sm.set_array([])

cbar = plt.colorbar(sm, ax=ax)
cbar.set_label('FLT_TOT_1', rotation=270, labelpad=15)
plt.savefig("figure9.png", bbox_inches='tight')

plt.show()

print(f"Explained variance ratio by PC1: {pca.explained_variance_ratio_[0]:.2f}")
print(f"Explained variance ratio by PC2: {pca.explained_variance_ratio_[1]:.2f}")
print(f"Total explained variance by 2 components: {pca.explained_variance_ratio_.sum():.2f}")
```

### Interpretation of Graphs:

1.  **Distribution of Total Daily IFR Movements (FLT_TOT_1)**:
    *   **What it shows:** This histogram displays the frequency of different total daily flight counts.
    *   **Interpretation:** The distribution is highly **right-skewed**, meaning a vast majority of daily airport observations have very low flight numbers (close to zero), while a much smaller number of observations show very high flight counts. The peak is concentrated at the lower end, indicating many days with minimal activity, likely from smaller airports or during periods of low travel demand (e.g., certain pandemic phases).
    *   **Implication for Modeling:** This severe skewness is problematic for many linear regression models, which assume normally distributed residuals. It suggests that a **transformation of the target variable (`FLT_TOT_1`)**, such as a logarithmic transformation, will be crucial to make the distribution more symmetrical and improve model performance.

2.  **Distribution of Log-Transformed Total Daily IFR Movements (log1p(FLT_TOT_1))**:
    *   **What it shows:** This histogram displays the frequency of log-transformed total daily flight counts (using `log1p` to handle zero values gracefully).
    *   **Interpretation:** After the `log1p` transformation, the distribution becomes significantly more symmetrical and spread out. While not perfectly Gaussian, it is much closer to a normal distribution, with peaks around 1.0-2.0 and 4.0-5.0 (corresponding to roughly 1.7-6.4 and 53.6-147.4 flights, respectively). The transformed variable is far more suitable for models that prefer or perform better with less skewed targets.
    *   **Implication for Modeling:** This plot confirms that `log1p` transformation is an **essential preprocessing step for the target variable**. The model will predict the log-transformed traffic, and the prediction will then be exponentiated back to the original scale (`np.expm1(predicted_log_value)`).

3.  **Total Flights by Airport (Top 10 by Average Traffic)**:
    *   **What it shows:** This box plot illustrates the distribution of `FLT_TOT_1` for the top 10 airports with the highest average traffic.
    *   **Interpretation:** There's a wide range in typical daily traffic even among the busiest airports. For instance, EDDF (Frankfurt) and LFPG (Paris CDG) show higher median daily movements and broader distributions compared to others like LTFM (Istanbul) or LTBQ (Izmir), which despite being "top" by average, have lower medians. All airports exhibit a significant number of **outliers** (circles above the whiskers), representing days with exceptionally high traffic. The lower whisker touching zero for several airports (e.g., LEBL, LTAI) indicates days with very few or no flights, even for relatively busy hubs.
    *   **Implication for Modeling:** `APT_ICAO` is a **highly critical categorical feature**, as it directly accounts for the vast differences in airport-specific traffic baselines and variability. The model must learn these distinct patterns for each airport. The presence of outliers suggests that models robust to extreme values, or careful outlier handling, might be beneficial.

4.  **Total Flights by Pandemic Phase**:
    *   **What it shows:** This box plot groups `FLT_TOT_1` by the engineered `pandemic_phase` categorical variable, ordered chronologically.
    *   **Interpretation:** The plot vividly demonstrates the catastrophic impact of the pandemic. Traffic sharply declines from "Pre-Pandemic (2016-2019)" levels to extremely low numbers during "Pandemic Peak Drop (Spring 2020)" and "Pandemic Continued Stagnation (Early 2021)". A gradual recovery is evident through "Pandemic Initial Recovery (Summer 2020)", "Pandemic Strong Recovery (Summer 2021)", and "Post-Pandemic Recovery (2022)". Even during the deepest troughs, high outliers (individual airports with significant traffic, possibly cargo or essential flights) persist.
    *   **Implication for Modeling:** The `pandemic_phase` feature is **paramount** for capturing the large, non-linear shifts in flight traffic caused by the global health crisis. This engineered feature explicitly encodes the unique, severe, and phased nature of the pandemic's influence, which raw `YEAR` or `MONTH_NUM` features alone cannot fully convey.

5.  **Traffic as Percentage of Pre-Pandemic Monthly Average by Phase**:
    *   **What it shows:** This box plot displays the `traffic_pct_of_pre_pandemic` (current traffic relative to its pre-pandemic monthly average) for each `pandemic_phase`.
    *   **Interpretation:** This is a powerful normalization. During the "Peak Drop" and "Continued Stagnation" phases, the median traffic was consistently near 0-20% of pre-pandemic levels, underscoring the severity. The recovery trend is clearly visible, with the median percentage gradually increasing towards 100% (or above) in the later phases. The most striking observation is the presence of numerous, extremely high outliers (e.g., values well over 1000% or even 4000%). These "super-outliers" likely represent very small airports that had almost no pre-pandemic traffic for a given month, but then handled a few flights during the pandemic, leading to a massive percentage increase, or perhaps specific cargo operations that surged.
    *   **Implication for Modeling:** This feature provides a **relative measure** of recovery, which can help the model generalize across airports of different sizes. While it's a derived feature from the target and a baseline, its potential for normalization makes it valuable. However, the extreme outliers in this *feature* itself (if used as an input) might require careful handling or a decision to cap its values to prevent undue influence on the model.

6.  **Total Flights by Airport Volume Category**:
    *   **What it shows:** This box plot groups `FLT_TOT_1` by the `airport_volume_category` (Small, Medium, Large, Very Large) derived from pre-pandemic average traffic.
    *   **Interpretation:** There's a clear and distinct separation of flight volumes across the different airport categories. "Very Large" airports consistently exhibit much higher median traffic and a wider range of activities compared to "Small" airports, which have very low traffic. This validates the effectiveness of this categorical feature.
    *   **Implication for Modeling:** This feature serves as a valuable **generalization** of `APT_ICAO`. It allows the model to learn patterns applicable to groups of similarly sized airports, which can improve generalization to new or less-frequently observed airports and provide a more robust understanding of the "scale" of an airport's operation.

7.  **Total Flights by Day of Week, Grouped by Airport Volume Category**:
    *   **What it shows:** This complex box plot visualizes the interaction between `day_of_week` and `airport_volume_category` on `FLT_TOT_1`.
    *   **Interpretation:**
        *   **Weekly Seasonality:** For all airport categories, weekdays (Mon-Fri) generally show higher traffic than weekends (Sat-Sun).
        *   **Interaction Effect:** This plot brilliantly highlights a significant **interaction**. The *magnitude* of the difference between weekday and weekend traffic is drastically more pronounced for "Very Large" and "Large" airports. For "Small" airports, the median traffic remains consistently low across all days of the week, with less variability. This suggests that the weekly rhythm of commercial air travel is primarily driven by large hubs.
    *   **Implication for Modeling:** This strong interaction effect means that the model cannot simply consider `day_of_week` and `airport_volume_category` independently. A good model must be able to learn that the impact of a weekend day is much more significant for a large airport than for a small one.

8.  **Average Total Flights by Year and Month (Heatmap)**:
    *   **What it shows:** A heatmap displaying the average `FLT_TOT_1` for each combination of `YEAR` and `MONTH_NUM`.
    *   **Interpretation:** This plot perfectly summarizes the multi-layered temporal patterns.
        *   **Pre-Pandemic Baseline (2016-2019):** Shows relatively consistent annual cycles, with higher averages in summer months (June-September) and dips in winter.
        *   **Pandemic Impact (2020-2021):** The dramatic drop in 2020 (especially April-May) is evident from the light-colored cells, indicating very low average traffic. 2021 shows a suppressed but gradually recovering pattern compared to 2020.
        *   **2022 Recovery:** The values for 2022 (up to May) are noticeably higher than corresponding months in 2021, indicating a strong recovery but not yet fully returning to pre-pandemic levels.
    *   **Implication for Modeling:** This heatmap confirms the critical importance of both `YEAR` and `MONTH_NUM` features, and especially their **interaction**, as the "meaning" of a month (e.g., April) changes drastically depending on the year due to the pandemic. The model needs to capture these strong annual, monthly, and event-driven patterns.

### Model Selection based on Insights:

The problem of predicting daily flight traffic is a **regression problem** characterized by a highly skewed target variable, strong time-series components (weekly, monthly, annual cycles), the profound, non-linear impact of the COVID-19 pandemic, and the presence of numerous high-cardinality categorical features with significant interaction effects.

Given these complexities, simple linear models will likely perform poorly without extensive manual feature engineering. More sophisticated models that can handle non-linearities and interactions are essential.

**1. Sophisticated Solutions (Tree-Based Ensemble Models):**

*   **CatBoost Regressor:** This would be my **top recommendation**. CatBoost is specifically designed to handle **categorical features natively**, meaning you can feed `APT_ICAO` (with its hundreds of unique values), `STATE_NAME`, `pandemic_phase`, `day_of_week`, `MONTH_NUM`, and `airport_volume_category` directly as categorical inputs without explicit one-hot or label encoding. This is a massive advantage given the dataset's structure. It automatically handles interactions between features and is robust to outliers and skewed distributions.
*   **XGBoost Regressor / LightGBM Regressor:** These are excellent, high-performance gradient boosting frameworks. They excel at capturing complex non-linear relationships and interactions. For these models, high-cardinality categorical features like `APT_ICAO` and `STATE_NAME` would ideally be handled with **Label Encoding** (often combined with `tree_method='hist'` for faster and more robust performance), or potentially **Target Encoding** (with careful cross-validation to prevent data leakage). Other categorical features can be Label Encoded.

For all tree-based models, the target variable (`FLT_TOT_1`) should be **log-transformed (`np.log1p`)** to stabilize variance and improve performance, with predictions then inverse-transformed (`np.expm1`).

**2. Sophisticated Solution (Deep Learning - Neural Networks):**

*   A **Neural Network with Embedding Layers** presents a powerful alternative. Numerical features would be standardized. For categorical features (especially high-cardinality ones like `APT_ICAO`, `STATE_NAME`), embedding layers can learn dense, low-dimensional representations. These embeddings can then be concatenated with numerical features and fed into dense layers. A well-designed neural network can capture very complex non-linear relationships and interactions.
*   **Consideration:** While potentially offering very high performance, building, training, and tuning a neural network can be more time-consuming and computationally intensive than gradient boosting models.

**3. Naive Solution (Linear Regression / Ridge Regression):**

*   **Linear Regression or Ridge Regression** could serve as a **baseline model**.
*   **Preprocessing for Linear Models:** This would require **extensive preprocessing**:
    *   **Log-transform `FLT_TOT_1`**.
    *   **One-Hot Encoding** for *all* categorical features, including `APT_ICAO` and `STATE_NAME`, which would result in a very high-dimensional and sparse feature matrix (hundreds of columns from `APT_ICAO` alone).
    *   All numerical features would need to be **standardized**.
    *   To capture the observed interactions (e.g., between `day_of_week` and `airport_volume_category`, or `YEAR` and `MONTH_NUM` for pandemic effect), explicit **interaction terms** would need to be manually engineered and added as new features.
*   **Limitations:** Despite the detailed feature engineering, a simple linear model would inherently struggle to model the sharp non-linear drops and recoveries due to the pandemic as effectively as tree-based or deep learning models, and its interpretability would be hampered by the vast number of one-hot encoded and interaction terms. Its performance would likely be significantly lower than the sophisticated models.

In conclusion, the data's characteristics, particularly the complex temporal dynamics (including the pandemic's phased impact), the high-cardinality categorical features, and significant interaction effects, strongly advocate for the use of **tree-based ensemble models (especially CatBoost)** or **deep learning with embedding layers**. These models are best equipped to learn the intricate patterns and provide robust predictions for daily flight traffic. The naive linear model would serve as a crucial benchmark, but is unlikely to be the optimal solution.

```python
import pandas as pd
import numpy as np
import datetime
import time
import pytz # Import pytz to handle timezones
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from catboost import CatBoostRegressor, Pool
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# --- Configuration ---
RANDOM_SEED = 42
# Data Period: Jan 2016 - May 2022 (77 months total)
# 80% of 77 months = ~61.6 months.
# So, data up to 2016-01-01 + 61 months = 2021-02-01.
# Make CHRONOLOGICAL_SPLIT_DATE timezone-aware (UTC) to match FLT_DATE
CHRONOLOGICAL_SPLIT_DATE = datetime.datetime(2021, 2, 1, tzinfo=pytz.UTC)

# Store results for all models
results = {}

# --- 1. Data Loading and Initial Cleaning ---
data_raw = pd.read_csv("flights.csv")

columns_to_drop_initial = [
    'FLT_DEP_IFR_2', 'FLT_ARR_IFR_2', 'FLT_TOT_IFR_2', # High NaNs, redundant
    'MONTH_MON',     # Redundant with MONTH_NUM
    'Pivot Label',   # Redundant with APT_ICAO and APT_NAME
    'APT_NAME'       # Redundant with APT_ICAO
]
data_cleaned = data_raw.drop(columns=columns_to_drop_initial)

# Convert FLT_DATE to datetime object for proper chronological split and feature engineering
# Ensure it's timezone-aware (UTC) as per the data description
data_cleaned['FLT_DATE'] = pd.to_datetime(data_cleaned['FLT_DATE'])

# --- 2. Feature Engineering ---

# Time-based features
data_cleaned['day_of_week'] = data_cleaned['FLT_DATE'].dt.dayofweek # Monday=0, Sunday=6
data_cleaned['day_of_month'] = data_cleaned['FLT_DATE'].dt.day
data_cleaned['day_of_year'] = data_cleaned['FLT_DATE'].dt.dayofyear
data_cleaned['week_of_year'] = data_cleaned['FLT_DATE'].dt.isocalendar().week.astype(int)
data_cleaned['quarter'] = data_cleaned['FLT_DATE'].dt.quarter
data_cleaned['is_weekend'] = (data_cleaned['FLT_DATE'].dt.dayofweek >= 5).astype(int)

# Pandemic Phase Feature (from previous discussion)
def assign_pandemic_phase(row):
    year = row['YEAR']
    month = row['MONTH_NUM']
    if year < 2020:
        return 'Pre-Pandemic (2016-2019)'
    elif year == 2020:
        if month >= 3 and month <= 5: # Initial severe drop
            return 'Pandemic Peak Drop (Spring 2020)'
        elif month >= 6 and month <= 8: # Slight summer recovery
            return 'Pandemic Initial Recovery (Summer 2020)'
        else: # Sep-Dec 2020: Second wave / Fall-Winter stagnation
            return 'Pandemic Second Wave (Fall/Winter 2020)'
    elif year == 2021:
        if month >= 1 and month <= 4: # Jan-April 2021: Continued stagnation
            return 'Pandemic Continued Stagnation (Early 2021)'
        elif month >= 5 and month <= 8: # May-Aug 2021: Stronger recovery
            return 'Pandemic Strong Recovery (Summer 2021)'
        else: # Sep-Dec 2021: Post-summer recovery continues
            return 'Pandemic Post-Summer Recovery (Late 2021)'
    elif year == 2022:
        return 'Post-Pandemic Recovery (2022)'
    return 'Other' # Fallback, though should not be hit with this data

data_cleaned['pandemic_phase'] = data_cleaned.apply(assign_pandemic_phase, axis=1)

# Pre-pandemic average traffic (by Airport and Month)
pre_pandemic_data = data_cleaned[data_cleaned['YEAR'] < 2020].copy()
avg_traffic_pre_pandemic_monthly = pre_pandemic_data.groupby(['APT_ICAO', 'MONTH_NUM'])['FLT_TOT_1'].mean().reset_index()
avg_traffic_pre_pandemic_monthly.rename(columns={'FLT_TOT_1': 'avg_pre_pandemic_monthly_traffic'}, inplace=True)
data_cleaned = pd.merge(data_cleaned, avg_traffic_pre_pandemic_monthly,
                        on=['APT_ICAO', 'MONTH_NUM'], how='left')

# Fix FutureWarning: Reassign the column instead of using inplace
data_cleaned['avg_pre_pandemic_monthly_traffic'] = data_cleaned['avg_pre_pandemic_monthly_traffic'].fillna(
    pre_pandemic_data['FLT_TOT_1'].mean()
)

# Traffic as Percentage of Pre-Pandemic Monthly Average
data_cleaned['traffic_pct_of_pre_pandemic'] = (
    data_cleaned['FLT_TOT_1'] / data_cleaned['avg_pre_pandemic_monthly_traffic']
) * 100
# Fix FutureWarning: Reassign the column instead of using inplace
data_cleaned['traffic_pct_of_pre_pandemic'] = data_cleaned['traffic_pct_of_pre_pandemic'].replace([np.inf, -np.inf], np.nan)
data_cleaned['traffic_pct_of_pre_pandemic'] = data_cleaned['traffic_pct_of_pre_pandemic'].fillna(0) # If avg was 0 and current is 0

# Airport Volume Category (based on overall pre-pandemic average)
avg_traffic_pre_pandemic_overall = pre_pandemic_data.groupby('APT_ICAO')['FLT_TOT_1'].mean().reset_index()
avg_traffic_pre_pandemic_overall.rename(columns={'FLT_TOT_1': 'avg_traffic_pre_pandemic_overall'}, inplace=True)
data_cleaned = pd.merge(data_cleaned, avg_traffic_pre_pandemic_overall,
                        on='APT_ICAO', how='left')
# Fix FutureWarning: Reassign the column instead of using inplace
data_cleaned['avg_traffic_pre_pandemic_overall'] = data_cleaned['avg_traffic_pre_pandemic_overall'].fillna(
    pre_pandemic_data['FLT_TOT_1'].mean()
)

data_cleaned['airport_volume_category'] = pd.qcut(
    data_cleaned['avg_traffic_pre_pandemic_overall'],
    q=4, labels=['Small', 'Medium', 'Large', 'Very Large'], duplicates='drop'
)

# --- 3. Define Features (X) and Target (y) and Chronological Data Split ---

# Target transformation: $y' = \log(1+y)$
# y_original remains for inverse transformation and original scale metrics
y_original = data_cleaned['FLT_TOT_1']
y_transformed = np.log1p(y_original)

# Features: exclude the target itself and its direct components
features_to_exclude_from_X = ['FLT_DEP_1', 'FLT_ARR_1', 'FLT_TOT_1']

# Perform the chronological split using 'FLT_DATE'
train_df = data_cleaned[data_cleaned['FLT_DATE'] < CHRONOLOGICAL_SPLIT_DATE].copy()
test_df = data_cleaned[data_cleaned['FLT_DATE'] >= CHRONOLOGICAL_SPLIT_DATE].copy()

# Drop 'FLT_DATE' from feature sets after splitting
X_train = train_df.drop(columns=features_to_exclude_from_X + ['FLT_DATE'])
y_train_original = train_df['FLT_TOT_1']
y_train_transformed = np.log1p(y_train_original)

X_test = test_df.drop(columns=features_to_exclude_from_X + ['FLT_DATE'])
y_test_original = test_df['FLT_TOT_1']
y_test_transformed = np.log1p(y_test_original)

# Define numerical and categorical features for different models
numerical_features = [
    'YEAR', 'day_of_month', 'day_of_year', 'week_of_year', 'is_weekend',
    'avg_pre_pandemic_monthly_traffic', 'avg_traffic_pre_pandemic_overall',
    'traffic_pct_of_pre_pandemic' # This feature can be highly volatile, consider capping if it causes issues
]
categorical_features = [
    'APT_ICAO', 'STATE_NAME', 'MONTH_NUM', 'day_of_week', 'quarter',
    'pandemic_phase', 'airport_volume_category'
]

# Ensure categorical features are of 'category' dtype for consistent handling by CatBoost
for col in categorical_features:
    if col in X_train.columns:
        X_train[col] = X_train[col].astype('category')
    if col in X_test.columns:
        X_test[col] = X_test[col].astype('category')

print(f"Train set shape: {X_train.shape}, Test set shape: {X_test.shape}")
print(f"Train/Test chronological split ratio: {len(X_train) / (len(X_train) + len(X_test)):.2f}")

# --- Helper function for metrics calculation ---
def calculate_metrics(y_true_transformed, y_pred_transformed, y_true_original, y_pred_original_safe):
    # Metrics on log-transformed scale
    r2_log = r2_score(y_true_transformed, y_pred_transformed)
    mae_log = mean_absolute_error(y_true_transformed, y_pred_transformed)
    rmse_log = np.sqrt(mean_squared_error(y_true_transformed, y_pred_transformed))

    # Metrics on original scale
    r2_orig = r2_score(y_true_original, y_pred_original_safe)
    mae_orig = mean_absolute_error(y_true_original, y_pred_original_safe)
    rmse_orig = np.sqrt(mean_squared_error(y_true_original, y_pred_original_safe))

    return {
        "metrics_log_scale": {"R2": r2_log, "MAE": mae_log, "RMSE": rmse_log},
        "metrics_original_scale": {"R2": r2_orig, "MAE": mae_orig, "RMSE": rmse_orig},
    }

def record_model_performance(model, model_name, X_train, y_train_transformed, y_train_original,
                             X_test, y_test_transformed, y_test_original, model_params=None):
    start_train_time = time.time()
    if model_name == "CatBoost":
        cat_features_indices = [X_train.columns.get_loc(col) for col in categorical_features if col in X_train.columns]
        model.fit(X_train, y_train_transformed, cat_features=cat_features_indices,
                  eval_set=(X_test, y_test_transformed), verbose=False)
    elif model_name == "XGBoost":
        model.fit(X_train, y_train_transformed,
                  eval_set=[(X_test, y_test_transformed)], verbose=False)
    else: # Linear Regression in a Pipeline
        model.fit(X_train, y_train_transformed)
    end_train_time = time.time()
    train_time_sec = end_train_time - start_train_time

    # Predict on train set
    start_predict_train_time = time.time()
    y_pred_train_transformed = model.predict(X_train)
    end_predict_train_time = time.time()
    predict_train_time_sec = end_predict_train_time - start_predict_train_time
    # Ensure predictions are non-negative after inverse log transformation
    y_pred_train_original_safe = np.expm1(y_pred_train_transformed)
    y_pred_train_original_safe[y_pred_train_original_safe < 0] = 0 # Cap at 0
    train_metrics = calculate_metrics(y_train_transformed, y_pred_train_transformed,
                                      y_train_original, y_pred_train_original_safe)

    # Predict on test set
    start_predict_test_time = time.time()
    y_pred_test_transformed = model.predict(X_test)
    end_predict_test_time = time.time()
    predict_test_time_sec = end_predict_test_time - start_predict_test_time
    # Ensure predictions are non-negative after inverse log transformation
    y_pred_test_original_safe = np.expm1(y_pred_test_transformed)
    y_pred_test_original_safe[y_pred_test_original_safe < 0] = 0 # Cap at 0
    test_metrics = calculate_metrics(y_test_transformed, y_pred_test_transformed,
                                     y_test_original, y_pred_test_original_safe)

    results[model_name] = {
        "model_params": model_params if model_params is not None else {},
        "train_time_sec": train_time_sec,
        "predict_time_sec_train": predict_train_time_sec,
        "predict_time_sec_test": predict_test_time_sec,
        **train_metrics, # Unpack train_metrics into top level
        "metrics_log_scale_test": test_metrics["metrics_log_scale"],
        "metrics_original_scale_test": test_metrics["metrics_original_scale"]
    }

# --- Model 1: CatBoost Regressor (Sophisticated Solution 1) ---
print("\n--- Training CatBoost Regressor ---")
cat_params = {
    'iterations': 1000,
    'learning_rate': 0.05,
    'depth': 8,
    'loss_function': 'RMSE',
    'eval_metric': 'RMSE',
    'random_seed': RANDOM_SEED,
    'verbose': 0,
    'early_stopping_rounds': 50
}
cat_model = CatBoostRegressor(**cat_params)
record_model_performance(cat_model, "CatBoost", X_train, y_train_transformed, y_train_original,
                         X_test, y_test_transformed, y_test_original, cat_params)
print("CatBoost Training Complete.")


# --- Model 2: XGBoost Regressor (Sophisticated Solution 2) ---
print("\n--- Training XGBoost Regressor ---")
# Preprocessing for XGBoost: Label Encoding for categorical features
X_train_xgb = X_train.copy()
X_test_xgb = X_test.copy()

label_encoders = {}
for col in categorical_features:
    if col in X_train_xgb.columns:
        le = LabelEncoder()
        # Fit on train data only to avoid data leakage
        X_train_xgb[col] = le.fit_transform(X_train_xgb[col])
        # Transform test data using fitted encoder. Handle unseen labels in test set.
        # Create a mapping for existing classes
        le_mapping = {label: idx for idx, label in enumerate(le.classes_)}
        # Apply mapping, assign a new unseen value (e.g., max_idx + 1) for new categories
        # Or, just use -1 or 0 for unseen, depends on XGBoost's handling. Using -1 here for simplicity.
        X_test_xgb[col] = X_test_xgb[col].apply(lambda x: le_mapping.get(x, -1))
        # Ensure integer type for XGBoost
        X_test_xgb[col] = X_test_xgb[col].astype(int)


xgb_params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'n_estimators': 1000,
    'learning_rate': 0.05,
    'max_depth': 8,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'random_state': RANDOM_SEED,
    'n_jobs': -1,
    'tree_method': 'hist', # For performance on large datasets
    'early_stopping_rounds': 50
}
xgb_model = xgb.XGBRegressor(**xgb_params)
record_model_performance(xgb_model, "XGBoost", X_train_xgb, y_train_transformed, y_train_original,
                         X_test_xgb, y_test_transformed, y_test_original, xgb_params)
print("XGBoost Training Complete.")


# --- Model 3: Linear Regression (Naive Solution) ---
print("\n--- Training Linear Regression (Naive) ---")
# Preprocessing for Linear Regression: StandardScaler for numerical, OneHotEncoder for categoricals
preprocessor_lr = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features) # handle_unknown='ignore' for unseen test categories
    ],
    remainder='passthrough'
)

lr_model = Pipeline(steps=[('preprocessor', preprocessor_lr),
                           ('regressor', LinearRegression(n_jobs=-1))])
record_model_performance(lr_model, "LinearRegression", X_train, y_train_transformed, y_train_original,
                         X_test, y_test_transformed, y_test_original)
print("Linear Regression Training Complete.")

# --- Print Final Results ---
print("\n--- Model Training & Performance Summary ---")
for model_name, model_data in results.items():
    print(f"\nModel: {model_name}")
    print(f"  Training Time: {model_data['train_time_sec']:.2f} seconds")
    print(f"  Prediction Time (Train Set): {model_data['predict_time_sec_train']:.4f} seconds")
    print(f"  Prediction Time (Test Set): {model_data['predict_time_sec_test']:.4f} seconds")

    print("  Train Metrics (Log Scale):")
    for metric, value in model_data['metrics_log_scale'].items():
        print(f"    {metric}: {value:.4f}")
    print("  Train Metrics (Original Scale):")
    for metric, value in model_data['metrics_original_scale'].items():
        print(f"    {metric}: {value:.4f}")

    print("  Test Metrics (Log Scale):")
    for metric, value in model_data['metrics_log_scale_test'].items():
        print(f"    {metric}: {value:.4f}")
    print("  Test Metrics (Original Scale):")
    for metric, value in model_data['metrics_original_scale_test'].items():
        print(f"    {metric}: {value:.4f}")
```

Data leakage occurs when information from outside the training data, or information that would not be available at prediction time, is used to train the model. This causes the model to appear to perform exceptionally well during evaluation, but it will fail miserably on genuinely new, unseen data in a real-world deployment.

There are several common forms of data leakage:

1.  **Target Leakage (Direct):** When the target variable itself, or a feature directly computed from the target variable that is not available at prediction time, is included in the features.
2.  **Train-Test Contamination:** When information from the test set "leaks" into the training process (e.g., scaling or imputation using the entire dataset before splitting). We addressed this with chronological split and proper preprocessing pipelines.
3.  **Future Leakage:** When future information inadvertently influences current predictions (e.g., using statistics calculated *after* the prediction point). Our chronological split largely mitigates this.

### The Culprit: `traffic_pct_of_pre_pandemic` Feature

Let's re-examine how the `traffic_pct_of_pre_pandemic` feature was constructed:

$$ \text{traffic\_pct\_of\_pre\_pandemic} = \frac{\text{FLT\_TOT\_1}}{\text{avg\_pre\_pandemic\_monthly\_traffic}} \times 100 $$

Here's why this is a direct form of **target leakage**:

*   **$y = \text{FLT\_TOT\_1}$**: This is our target variable, the number of flights for a given day at a given airport, which we are trying to *predict*.
*   **$\text{avg\_pre\_pandemic\_monthly\_traffic}$**: This feature is calculated as the average `FLT_TOT_1` for a specific airport and month *only from pre-2020 data*. This specific feature on its own is perfectly valid as it represents a historical baseline that would be known at prediction time.

However, when we define:
$$ F_{\text{leak}} = \text{traffic\_pct\_of\_pre\_pandemic} $$
and then include $F_{\text{leak}}$ as a feature in our input matrix $X$, the model is implicitly given knowledge of the target variable $y$.

**Mathematically, the model can essentially learn the following relationship:**

Given the feature $F_{\text{leak}}$ and $F_{\text{avg}} = \text{avg\_pre\_pandemic\_monthly\_traffic}$, the model can directly infer the target $y$:

$$ y \approx F_{\text{leak}} \times \frac{F_{\text{avg}}}{100} $$

Since $F_{\text{avg}}$ is a known constant (for a given airport and month) and $F_{\text{leak}}$ directly incorporates the target $y$, the model doesn't need to "learn" anything complex. It just needs to figure out this simple multiplication.

This is akin to predicting someone's age on a specific day, given their birthday, by providing a feature that is "number of days since their birth". The model would achieve near 100% accuracy because the information is already fully contained in the input feature.

### Why the High Accuracy?

Because the model has a feature (`traffic_pct_of_pre_pandemic`) that is a direct algebraic transformation of the target variable, it can virtually perfectly reconstruct the target during training and even on the test set (as long as `avg_pre_pandemic_monthly_traffic` is not zero and is consistent between train and test, which it is because it's calculated from fixed historical data).

This explains:
*   The incredibly high R2 values (0.999+).
*   The extremely low MAE and RMSE values (e.g., 1.37 flights error on average).
*   Why CatBoost and XGBoost (which are powerful non-linear learners) exploited this relationship so effectively.
*   Even Linear Regression showed a high R2 on the log-transformed scale (0.9294) because the linear relationship $y' \approx \log(1 + F_{\text{leak}} \times \frac{F_{\text{avg}}}{100})$ is still relatively simple to model after transformation, although its inverse transformation caused issues with negative predictions which lead to the catastrophic R2 on the original scale.

### Immediate Action Required: Remove the Leaking Feature

To properly evaluate the models and ensure they learn meaningful patterns that will generalize to truly unseen future data, the `traffic_pct_of_pre_pandemic` feature **must be removed** from the input feature set ($X$).

Once this feature is removed, the models will no longer have a "cheat sheet" and will be forced to learn the complex, non-linear relationships between the remaining independent variables (year, month, day of week, airport, state, pandemic phase, etc.) and the daily flight traffic. The R2 values will drop significantly (to more realistic levels), and the MAE/RMSE will increase, but the evaluation will then be a true reflection of the models' predictive power.

After removing this feature, we can re-train the models and expect more realistic (and still hopefully good) performance metrics.

```python
import pandas as pd
import numpy as np
import datetime
import time
import pytz # Import pytz to handle timezones
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from catboost import CatBoostRegressor, Pool
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# --- Configuration ---
RANDOM_SEED = 42
CHRONOLOGICAL_SPLIT_DATE = datetime.datetime(2021, 2, 1, tzinfo=pytz.UTC)

# Store results for all models
results = {}

# --- 1. Data Loading and Initial Cleaning ---
data_raw = pd.read_csv("flights.csv")

columns_to_drop_initial = [
    'FLT_DEP_IFR_2', 'FLT_ARR_IFR_2', 'FLT_TOT_IFR_2', # High NaNs, redundant
    'MONTH_MON',     # Redundant with MONTH_NUM
    'Pivot Label',   # Redundant with APT_ICAO and APT_NAME
    'APT_NAME'       # Redundant with APT_ICAO
]
data_cleaned = data_raw.drop(columns=columns_to_drop_initial)

data_cleaned['FLT_DATE'] = pd.to_datetime(data_cleaned['FLT_DATE'])

# --- 2. Feature Engineering ---

# Time-based features
data_cleaned['day_of_week'] = data_cleaned['FLT_DATE'].dt.dayofweek
data_cleaned['day_of_month'] = data_cleaned['FLT_DATE'].dt.day
data_cleaned['day_of_year'] = data_cleaned['FLT_DATE'].dt.dayofyear
data_cleaned['week_of_year'] = data_cleaned['FLT_DATE'].dt.isocalendar().week.astype(int)
data_cleaned['quarter'] = data_cleaned['FLT_DATE'].dt.quarter
data_cleaned['is_weekend'] = (data_cleaned['FLT_DATE'].dt.dayofweek >= 5).astype(int)

# Pandemic Phase Feature
def assign_pandemic_phase(row):
    year = row['YEAR']
    month = row['MONTH_NUM']
    if year < 2020:
        return 'Pre-Pandemic (2016-2019)'
    elif year == 2020:
        if month >= 3 and month <= 5:
            return 'Pandemic Peak Drop (Spring 2020)'
        elif month >= 6 and month <= 8:
            return 'Pandemic Initial Recovery (Summer 2020)'
        else:
            return 'Pandemic Second Wave (Fall/Winter 2020)'
    elif year == 2021:
        if month >= 1 and month <= 4:
            return 'Pandemic Continued Stagnation (Early 2021)'
        elif month >= 5 and month <= 8:
            return 'Pandemic Strong Recovery (Summer 2021)'
        else:
            return 'Pandemic Post-Summer Recovery (Late 2021)'
    elif year == 2022:
        return 'Post-Pandemic Recovery (2022)'
    return 'Other'

data_cleaned['pandemic_phase'] = data_cleaned.apply(assign_pandemic_phase, axis=1)

# Pre-pandemic average traffic (by Airport and Month)
pre_pandemic_data = data_cleaned[data_cleaned['YEAR'] < 2020].copy()
avg_traffic_pre_pandemic_monthly = pre_pandemic_data.groupby(['APT_ICAO', 'MONTH_NUM'])['FLT_TOT_1'].mean().reset_index()
avg_traffic_pre_pandemic_monthly.rename(columns={'FLT_TOT_1': 'avg_pre_pandemic_monthly_traffic'}, inplace=True)
data_cleaned = pd.merge(data_cleaned, avg_traffic_pre_pandemic_monthly,
                        on=['APT_ICAO', 'MONTH_NUM'], how='left')
data_cleaned['avg_pre_pandemic_monthly_traffic'] = data_cleaned['avg_pre_pandemic_monthly_traffic'].fillna(
    pre_pandemic_data['FLT_TOT_1'].mean()
)

# Traffic as Percentage of Pre-Pandemic Monthly Average (THIS IS THE LEAKING FEATURE IF USED IN X)
data_cleaned['traffic_pct_of_pre_pandemic'] = (
    data_cleaned['FLT_TOT_1'] / data_cleaned['avg_pre_pandemic_monthly_traffic']
) * 100
data_cleaned['traffic_pct_of_pre_pandemic'] = data_cleaned['traffic_pct_of_pre_pandemic'].replace([np.inf, -np.inf], np.nan)
data_cleaned['traffic_pct_of_pre_pandemic'] = data_cleaned['traffic_pct_of_pre_pandemic'].fillna(0)

# Airport Volume Category (based on overall pre-pandemic average)
avg_traffic_pre_pandemic_overall = pre_pandemic_data.groupby('APT_ICAO')['FLT_TOT_1'].mean().reset_index()
avg_traffic_pre_pandemic_overall.rename(columns={'FLT_TOT_1': 'avg_traffic_pre_pandemic_overall'}, inplace=True)
data_cleaned = pd.merge(data_cleaned, avg_traffic_pre_pandemic_overall,
                        on='APT_ICAO', how='left')
data_cleaned['avg_traffic_pre_pandemic_overall'] = data_cleaned['avg_traffic_pre_pandemic_overall'].fillna(
    pre_pandemic_data['FLT_TOT_1'].mean()
)
data_cleaned['airport_volume_category'] = pd.qcut(
    data_cleaned['avg_traffic_pre_pandemic_overall'],
    q=4, labels=['Small', 'Medium', 'Large', 'Very Large'], duplicates='drop'
)

# --- 3. Define Features (X) and Target (y) and Chronological Data Split ---

y_original = data_cleaned['FLT_TOT_1']
y_transformed = np.log1p(y_original)

# --- CRITICAL FIX: Define final features to be explicitly selected for X ---
numerical_features = [
    'YEAR', 'day_of_month', 'day_of_year', 'week_of_year', 'is_weekend',
    'avg_pre_pandemic_monthly_traffic', 'avg_traffic_pre_pandemic_overall'
    # 'traffic_pct_of_pre_pandemic' is INTENTIONALLY EXCLUDED to prevent leakage.
]
categorical_features = [
    'APT_ICAO', 'STATE_NAME', 'MONTH_NUM', 'day_of_week', 'quarter',
    'pandemic_phase', 'airport_volume_category'
]

# Combine all desired features into one list
final_model_features = numerical_features + categorical_features

# Perform the chronological split using 'FLT_DATE'
train_df = data_cleaned[data_cleaned['FLT_DATE'] < CHRONOLOGICAL_SPLIT_DATE].copy()
test_df = data_cleaned[data_cleaned['FLT_DATE'] >= CHRONOLOGICAL_SPLIT_DATE].copy()

# Select only the `final_model_features` for X_train and X_test
X_train = train_df[final_model_features].copy()
y_train_original = train_df['FLT_TOT_1'].copy()
y_train_transformed = np.log1p(y_train_original)

X_test = test_df[final_model_features].copy()
y_test_original = test_df['FLT_TOT_1'].copy()
y_test_transformed = np.log1p(y_test_original)

# Ensure categorical features are of 'category' dtype for consistent handling by CatBoost
for col in categorical_features:
    if col in X_train.columns:
        X_train[col] = X_train[col].astype('category')
    if col in X_test.columns:
        X_test[col] = X_test[col].astype('category')

print("X_train columns after fixing data leakage:")
print(X_train.columns.tolist())
print(f"Train set shape: {X_train.shape}, Test set shape: {X_test.shape}")
print(f"Train/Test chronological split ratio: {len(X_train) / (len(X_train) + len(X_test)):.2f}")


# --- Helper function for metrics calculation ---
def calculate_metrics(y_true_transformed, y_pred_transformed, y_true_original, y_pred_original_safe):
    # Metrics on log-transformed scale
    r2_log = r2_score(y_true_transformed, y_pred_transformed)
    mae_log = mean_absolute_error(y_true_transformed, y_pred_transformed)
    rmse_log = np.sqrt(mean_squared_error(y_true_transformed, y_pred_transformed))

    # Metrics on original scale
    r2_orig = r2_score(y_true_original, y_pred_original_safe)
    mae_orig = mean_absolute_error(y_true_original, y_pred_original_safe)
    rmse_orig = np.sqrt(mean_squared_error(y_true_original, y_pred_original_safe))

    return {
        "metrics_log_scale": {"R2": r2_log, "MAE": mae_log, "RMSE": rmse_log},
        "metrics_original_scale": {"R2": r2_orig, "MAE": mae_orig, "RMSE": rmse_orig},
    }

def record_model_performance(model, model_name, X_train, y_train_transformed, y_train_original,
                             X_test, y_test_transformed, y_test_original, model_params=None):
    start_train_time = time.time()
    if model_name == "CatBoost":
        cat_features_indices = [X_train.columns.get_loc(col) for col in categorical_features if col in X_train.columns]
        model.fit(X_train, y_train_transformed, cat_features=cat_features_indices,
                  eval_set=(X_test, y_test_transformed), verbose=False)
    elif model_name == "XGBoost":
        model.fit(X_train, y_train_transformed,
                  eval_set=[(X_test, y_test_transformed)], verbose=False)
    else: # Linear Regression in a Pipeline
        model.fit(X_train, y_train_transformed)
    end_train_time = time.time()
    train_time_sec = end_train_time - start_train_time

    # Predict on train set
    start_predict_train_time = time.time()
    y_pred_train_transformed = model.predict(X_train)
    end_predict_train_time = time.time()
    predict_train_time_sec = end_predict_train_time - start_predict_train_time
    y_pred_train_original_safe = np.expm1(y_pred_train_transformed)
    y_pred_train_original_safe[y_pred_train_original_safe < 0] = 0
    train_metrics = calculate_metrics(y_train_transformed, y_pred_train_transformed,
                                      y_train_original, y_pred_train_original_safe)

    # Predict on test set
    start_predict_test_time = time.time()
    y_pred_test_transformed = model.predict(X_test)
    end_predict_test_time = time.time()
    predict_test_time_sec = end_predict_test_time - start_predict_test_time
    y_pred_test_original_safe = np.expm1(y_pred_test_transformed)
    y_pred_test_original_safe[y_pred_test_original_safe < 0] = 0
    test_metrics = calculate_metrics(y_test_transformed, y_pred_test_transformed,
                                     y_test_original, y_pred_test_original_safe)

    results[model_name] = {
        "model_params": model_params if model_params is not None else {},
        "train_time_sec": train_time_sec,
        "predict_time_sec_train": predict_train_time_sec,
        "predict_time_sec_test": predict_test_time_sec,
        **train_metrics,
        "metrics_log_scale_test": test_metrics["metrics_log_scale"],
        "metrics_original_scale_test": test_metrics["metrics_original_scale"]
    }

# --- Model 1: CatBoost Regressor (Sophisticated Solution 1) ---
print("\n--- Training CatBoost Regressor ---")
cat_params = {
    'iterations': 1000,
    'learning_rate': 0.05,
    'depth': 8,
    'loss_function': 'RMSE',
    'eval_metric': 'RMSE',
    'random_seed': RANDOM_SEED,
    'verbose': 0,
    'early_stopping_rounds': 50
}
cat_model = CatBoostRegressor(**cat_params)
record_model_performance(cat_model, "CatBoost", X_train, y_train_transformed, y_train_original,
                         X_test, y_test_transformed, y_test_original, cat_params)
print("CatBoost Training Complete.")


# --- Model 2: XGBoost Regressor (Sophisticated Solution 2) ---
print("\n--- Training XGBoost Regressor ---")
# Preprocessing for XGBoost: Label Encoding for categorical features
X_train_xgb = X_train.copy()
X_test_xgb = X_test.copy()

label_encoders = {}
for col in categorical_features:
    if col in X_train_xgb.columns:
        le = LabelEncoder()
        X_train_xgb[col] = le.fit_transform(X_train_xgb[col])
        le_mapping = {label: idx for idx, label in enumerate(le.classes_)}
        X_test_xgb[col] = X_test_xgb[col].apply(lambda x: le_mapping.get(x, -1))
        X_test_xgb[col] = X_test_xgb[col].astype(int)

xgb_params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'n_estimators': 1000,
    'learning_rate': 0.05,
    'max_depth': 8,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'random_state': RANDOM_SEED,
    'n_jobs': -1,
    'tree_method': 'hist',
    'early_stopping_rounds': 50
}
xgb_model = xgb.XGBRegressor(**xgb_params)
record_model_performance(xgb_model, "XGBoost", X_train_xgb, y_train_transformed, y_train_original,
                         X_test_xgb, y_test_transformed, y_test_original, xgb_params)
print("XGBoost Training Complete.")


# --- Model 3: Linear Regression (Naive Solution) ---
print("\n--- Training Linear Regression (Naive) ---")
# Preprocessing for Linear Regression: StandardScaler for numerical, OneHotEncoder for categoricals
preprocessor_lr = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough'
)

lr_model = Pipeline(steps=[('preprocessor', preprocessor_lr),
                           ('regressor', LinearRegression(n_jobs=-1))])
record_model_performance(lr_model, "LinearRegression", X_train, y_train_transformed, y_train_original,
                         X_test, y_test_transformed, y_test_original)
print("Linear Regression Training Complete.")

# --- Print Final Results ---
print("\n--- Model Training & Performance Summary ---")
for model_name, model_data in results.items():
    print(f"\nModel: {model_name}")
    print(f"  Training Time: {model_data['train_time_sec']:.2f} seconds")
    print(f"  Prediction Time (Train Set): {model_data['predict_time_sec_train']:.4f} seconds")
    print(f"  Prediction Time (Test Set): {model_data['predict_time_sec_test']:.4f} seconds")

    print("  Train Metrics (Log Scale):")
    for metric, value in model_data['metrics_log_scale'].items():
        print(f"    {metric}: {value:.4f}")
    print("  Train Metrics (Original Scale):")
    for metric, value in model_data['metrics_original_scale'].items():
        print(f"    {metric}: {value:.4f}")

    print("  Test Metrics (Log Scale):")
    for metric, value in model_data['metrics_log_scale_test'].items():
        print(f"    {metric}: {value:.4f}")
    print("  Test Metrics (Original Scale):")
    for metric, value in model_data['metrics_original_scale_test'].items():
        print(f"    {metric}: {value:.4f}")
```

```python
# import pandas as pd
import numpy as np
import datetime
import time
import pytz
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from catboost import CatBoostRegressor, Pool
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV

# --- Configuration ---
RANDOM_SEED = 42
CHRONOLOGICAL_SPLIT_DATE = datetime.datetime(2021, 2, 1, tzinfo=pytz.UTC)

# Store final results for all models (including tuned ones)
results = {}
tuned_results = {} # To store results of the tuning process

# --- 1. Data Loading and Initial Cleaning ---
data_raw = pd.read_csv("flights.csv")

columns_to_drop_initial = [
    'FLT_DEP_IFR_2', 'FLT_ARR_IFR_2', 'FLT_TOT_IFR_2',
    'MONTH_MON',
    'Pivot Label',
    'APT_NAME'
]
data_cleaned = data_raw.drop(columns=columns_to_drop_initial)
data_cleaned['FLT_DATE'] = pd.to_datetime(data_cleaned['FLT_DATE'])

# --- 2. Feature Engineering ---
data_cleaned['day_of_week'] = data_cleaned['FLT_DATE'].dt.dayofweek
data_cleaned['day_of_month'] = data_cleaned['FLT_DATE'].dt.day
data_cleaned['day_of_year'] = data_cleaned['FLT_DATE'].dt.dayofyear
data_cleaned['week_of_year'] = data_cleaned['FLT_DATE'].dt.isocalendar().week.astype(int)
data_cleaned['quarter'] = data_cleaned['FLT_DATE'].dt.quarter
data_cleaned['is_weekend'] = (data_cleaned['FLT_DATE'].dt.dayofweek >= 5).astype(int)

def assign_pandemic_phase(row):
    year = row['YEAR']
    month = row['MONTH_NUM']
    if year < 2020: return 'Pre-Pandemic (2016-2019)'
    elif year == 2020:
        if month >= 3 and month <= 5: return 'Pandemic Peak Drop (Spring 2020)'
        elif month >= 6 and month <= 8: return 'Pandemic Initial Recovery (Summer 2020)'
        else: return 'Pandemic Second Wave (Fall/Winter 2020)'
    elif year == 2021:
        if month >= 1 and month <= 4: return 'Pandemic Continued Stagnation (Early 2021)'
        elif month >= 5 and month <= 8: return 'Pandemic Strong Recovery (Summer 2021)'
        else: return 'Pandemic Post-Summer Recovery (Late 2021)'
    elif year == 2022: return 'Post-Pandemic Recovery (2022)'
    return 'Other'

data_cleaned['pandemic_phase'] = data_cleaned.apply(assign_pandemic_phase, axis=1)

pre_pandemic_data = data_cleaned[data_cleaned['YEAR'] < 2020].copy()
avg_traffic_pre_pandemic_monthly = pre_pandemic_data.groupby(['APT_ICAO', 'MONTH_NUM'])['FLT_TOT_1'].mean().reset_index()
avg_traffic_pre_pandemic_monthly.rename(columns={'FLT_TOT_1': 'avg_pre_pandemic_monthly_traffic'}, inplace=True)
data_cleaned = pd.merge(data_cleaned, avg_traffic_pre_pandemic_monthly, on=['APT_ICAO', 'MONTH_NUM'], how='left')
data_cleaned['avg_pre_pandemic_monthly_traffic'] = data_cleaned['avg_pre_pandemic_monthly_traffic'].fillna(
    pre_pandemic_data['FLT_TOT_1'].mean()
)

data_cleaned['traffic_pct_of_pre_pandemic'] = (
    data_cleaned['FLT_TOT_1'] / data_cleaned['avg_pre_pandemic_monthly_traffic']
) * 100
data_cleaned['traffic_pct_of_pre_pandemic'] = data_cleaned['traffic_pct_of_pre_pandemic'].replace([np.inf, -np.inf], np.nan)
data_cleaned['traffic_pct_of_pre_pandemic'] = data_cleaned['traffic_pct_of_pre_pandemic'].fillna(0)

avg_traffic_pre_pandemic_overall = pre_pandemic_data.groupby('APT_ICAO')['FLT_TOT_1'].mean().reset_index()
avg_traffic_pre_pandemic_overall.rename(columns={'FLT_TOT_1': 'avg_traffic_pre_pandemic_overall'}, inplace=True)
data_cleaned = pd.merge(data_cleaned, avg_traffic_pre_pandemic_overall, on='APT_ICAO', how='left')
data_cleaned['avg_traffic_pre_pandemic_overall'] = data_cleaned['avg_traffic_pre_pandemic_overall'].fillna(
    pre_pandemic_data['FLT_TOT_1'].mean()
)
data_cleaned['airport_volume_category'] = pd.qcut(
    data_cleaned['avg_traffic_pre_pandemic_overall'],
    q=4, labels=['Small', 'Medium', 'Large', 'Very Large'], duplicates='drop'
)

# --- 3. Define Features (X) and Target (y) and Chronological Data Split ---

y_original = data_cleaned['FLT_TOT_1']
y_transformed = np.log1p(y_original)

# --- CRITICAL: Define final features to be explicitly selected for X ---
numerical_features = [
    'YEAR', 'day_of_month', 'day_of_year', 'week_of_year', 'is_weekend',
    'avg_pre_pandemic_monthly_traffic', 'avg_traffic_pre_pandemic_overall'
    # 'traffic_pct_of_pre_pandemic' is EXCLUDED to prevent leakage.
]
categorical_features = [
    'APT_ICAO', 'STATE_NAME', 'MONTH_NUM', 'day_of_week', 'quarter',
    'pandemic_phase', 'airport_volume_category'
]

final_model_features = numerical_features + categorical_features

# Perform the chronological split
train_df = data_cleaned[data_cleaned['FLT_DATE'] < CHRONOLOGICAL_SPLIT_DATE].copy()
test_df = data_cleaned[data_cleaned['FLT_DATE'] >= CHRONOLOGICAL_SPLIT_DATE].copy()

X_train = train_df[final_model_features].copy()
y_train_original = train_df['FLT_TOT_1'].copy()
y_train_transformed = np.log1p(y_train_original)

X_test = test_df[final_model_features].copy()
y_test_original = test_df['FLT_TOT_1'].copy()
y_test_transformed = np.log1p(y_test_original)

for col in categorical_features:
    if col in X_train.columns:
        X_train[col] = X_train[col].astype('category')
    if col in X_test.columns:
        X_test[col] = X_test[col].astype('category')

print("X_train columns (after leakage fix):", X_train.columns.tolist())
print(f"Train set shape: {X_train.shape}, Test set shape: {X_test.shape}")
print(f"Train/Test chronological split ratio: {len(X_train) / (len(X_train) + len(X_test)):.2f}")


# --- Helper function for metrics calculation ---
def calculate_metrics(y_true_transformed, y_pred_transformed, y_true_original, y_pred_original_safe):
    r2_log = r2_score(y_true_transformed, y_pred_transformed)
    mae_log = mean_absolute_error(y_true_transformed, y_pred_transformed)
    rmse_log = np.sqrt(mean_squared_error(y_true_transformed, y_pred_transformed))

    r2_orig = r2_score(y_true_original, y_pred_original_safe)
    mae_orig = mean_absolute_error(y_true_original, y_pred_original_safe)
    rmse_orig = np.sqrt(mean_squared_error(y_true_original, y_pred_original_safe))

    return {
        "metrics_log_scale": {"R2": r2_log, "MAE": mae_log, "RMSE": rmse_orig}, # Note: RMSE_orig is correct
        "metrics_original_scale": {"R2": r2_orig, "MAE": mae_orig, "RMSE": rmse_orig},
    }

# Function to record final model performance on the hold-out test set
def record_final_model_performance(model, model_name, X_train_full, y_train_transformed_full, y_train_original_full,
                             X_test_full, y_test_transformed_full, y_test_original_full, model_params=None):
    start_train_time = time.time()
    if model_name.startswith("CatBoost"): # Check for CatBoost or CatBoost (Tuned)
        cat_features_indices = [X_train_full.columns.get_loc(col) for col in categorical_features if col in X_train_full.columns]
        model.fit(X_train_full, y_train_transformed_full, cat_features=cat_features_indices,
                  eval_set=(X_test_full, y_test_transformed_full), verbose=False)
    elif model_name.startswith("XGBoost"): # Check for XGBoost or XGBoost (Tuned)
        model.fit(X_train_full, y_train_transformed_full,
                  eval_set=[(X_test_full, y_test_transformed_full)], verbose=False)
    else: # Linear Regression in a Pipeline
        model.fit(X_train_full, y_train_transformed_full)
    end_train_time = time.time()
    train_time_sec = end_train_time - start_train_time

    # Predict on train set
    start_predict_train_time = time.time()
    y_pred_train_transformed = model.predict(X_train_full)
    end_predict_train_time = time.time()
    predict_train_time_sec = end_predict_train_time - start_predict_train_time
    y_pred_train_original_safe = np.expm1(y_pred_train_transformed)
    y_pred_train_original_safe[y_pred_train_original_safe < 0] = 0
    train_metrics = calculate_metrics(y_train_transformed_full, y_pred_train_transformed,
                                      y_train_original_full, y_pred_train_original_safe)

    # Predict on test set
    start_predict_test_time = time.time()
    y_pred_test_transformed = model.predict(X_test_full)
    end_predict_test_time = time.time()
    predict_test_time_sec = end_predict_test_time - start_predict_test_time
    y_pred_test_original_safe = np.expm1(y_pred_test_transformed)
    y_pred_test_original_safe[y_pred_test_original_safe < 0] = 0
    test_metrics = calculate_metrics(y_test_transformed_full, y_pred_test_transformed,
                                     y_test_original_full, y_pred_test_original_safe)

    results[model_name] = {
        "model_params": model_params if model_params is not None else {},
        "train_time_sec": train_time_sec,
        "predict_time_sec_train": predict_train_time_sec,
        "predict_time_sec_test": predict_test_time_sec,
        **train_metrics,
        "metrics_log_scale_test": test_metrics["metrics_log_scale"],
        "metrics_original_scale_test": test_metrics["metrics_original_scale"]
    }


# --- Hyperparameter Tuning ---

# TimeSeriesSplit for cross-validation
tscv = TimeSeriesSplit(n_splits=5) # Example: 5 splits

# --- CatBoost Tuning ---
print("\n--- Tuning CatBoost Regressor ---")
cat_param_grid = {
    'learning_rate': [0.03, 0.05],
    'depth': [7, 9],
    'l2_leaf_reg': [1, 3], # L2 regularization
    'subsample': [0.7, 0.9] # Bagging fraction
}
# Base parameters (iterations and early stopping are handled outside GridSearchCV for eval_set)
cat_base_params = {
    'iterations': 1000, # Max iterations
    'loss_function': 'RMSE',
    'eval_metric': 'RMSE',
    'random_seed': RANDOM_SEED,
    'verbose': 0,
    'early_stopping_rounds': 50 # Early stopping
}

best_cat_rmse = float('inf')
best_cat_params = None
cat_tuning_results = []

# Manual Grid Search with TimeSeriesSplit and Early Stopping
for lr in cat_param_grid['learning_rate']:
    for d in cat_param_grid['depth']:
        for l2 in cat_param_grid['l2_leaf_reg']:
            for ss in cat_param_grid['subsample']:
                current_params = {**cat_base_params, 'learning_rate': lr, 'depth': d, 'l2_leaf_reg': l2, 'subsample': ss}
                fold_rmses = []
                print(f"  Testing CatBoost with params: {current_params}")
                
                for fold, (train_idx, val_idx) in enumerate(tscv.split(X_train)): # Split X_train (DataFrame)
                    X_fold_train_df, X_fold_val_df = X_train.iloc[train_idx], X_train.iloc[val_idx]
                    y_fold_train, y_fold_val = y_train_transformed.iloc[train_idx], y_train_transformed.iloc[val_idx]

                    # Create CatBoost Pool for the current fold
                    train_pool_fold = Pool(X_fold_train_df, y_fold_train, cat_features=categorical_features)
                    val_pool_fold = Pool(X_fold_val_df, y_fold_val, cat_features=categorical_features)

                    model = CatBoostRegressor(**current_params)
                    # Pass the Pool objects directly to fit
                    model.fit(train_pool_fold, eval_set=val_pool_fold, verbose=0)
                    
                    best_iteration_rmse = model.get_best_score()['validation']['RMSE']
                    fold_rmses.append(best_iteration_rmse)
                    
                avg_rmse = np.mean(fold_rmses)
                cat_tuning_results.append({'params': current_params, 'avg_rmse': avg_rmse})

                if avg_rmse < best_cat_rmse:
                    best_cat_rmse = avg_rmse
                    best_cat_params = current_params
                print(f"  Avg RMSE for this combo: {avg_rmse:.4f}")

print("\n--- CatBoost Tuning Complete ---")
print(f"Best CatBoost RMSE: {best_cat_rmse:.4f}")
print(f"Best CatBoost Parameters: {best_cat_params}")

# Train final CatBoost model with best parameters on the full training set
print("\n--- Training Final CatBoost Model ---")
final_cat_model = CatBoostRegressor(**best_cat_params)
record_final_model_performance(final_cat_model, "CatBoost (Tuned)", X_train, y_train_transformed, y_train_original,
                               X_test, y_test_transformed, y_test_original, best_cat_params)
print("Final CatBoost Model Training Complete.")


# --- XGBoost Tuning ---
print("\n--- Tuning XGBoost Regressor ---")
xgb_param_grid = {
    'learning_rate': [0.03, 0.05],
    'max_depth': [7, 9],
    'subsample': [0.7, 0.9],
    'colsample_bytree': [0.7, 0.9]
}
xgb_base_params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'n_estimators': 1000,
    'random_state': RANDOM_SEED,
    'n_jobs': -1,
    'tree_method': 'hist',
    'early_stopping_rounds': 50
}

best_xgb_rmse = float('inf')
best_xgb_params = None
xgb_tuning_results = []

# Preprocessing for XGBoost during tuning (Label Encoding)
X_train_xgb_tuned = X_train.copy()
# X_test_xgb_tuned is used for final performance recording, not in tuning loop
X_test_xgb_tuned = X_test.copy()

# Apply Label Encoding for XGBoost tuning
# This needs to be done on the full X_train before splitting for CV,
# and then applied to X_test_xgb_tuned for the final prediction
xgb_label_encoders = {}
for col in categorical_features:
    if col in X_train_xgb_tuned.columns:
        le = LabelEncoder()
        X_train_xgb_tuned[col] = le.fit_transform(X_train_xgb_tuned[col])
        xgb_label_encoders[col] = le
        # Apply the same encoder to X_test_xgb_tuned for consistency
        le_mapping = {label: idx for idx, label in enumerate(le.classes_)}
        X_test_xgb_tuned[col] = X_test_xgb_tuned[col].apply(lambda x: le_mapping.get(x, -1))
        X_test_xgb_tuned[col] = X_test_xgb_tuned[col].astype(int)


# Manual Grid Search with TimeSeriesSplit and Early Stopping
for lr in xgb_param_grid['learning_rate']:
    for md in xgb_param_grid['max_depth']:
        for ss in xgb_param_grid['subsample']:
            for cbt in xgb_param_grid['colsample_bytree']:
                current_params = {**xgb_base_params, 'learning_rate': lr, 'max_depth': md, 'subsample': ss, 'colsample_bytree': cbt}
                fold_rmses = []
                print(f"  Testing XGBoost with params: {current_params}")

                for fold, (train_idx, val_idx) in enumerate(tscv.split(X_train_xgb_tuned)):
                    X_fold_train, X_fold_val = X_train_xgb_tuned.iloc[train_idx], X_train_xgb_tuned.iloc[val_idx]
                    y_fold_train, y_fold_val = y_train_transformed.iloc[train_idx], y_train_transformed.iloc[val_idx]

                    model = xgb.XGBRegressor(**current_params)
                    model.fit(X_fold_train, y_fold_train, eval_set=[(X_fold_val, y_fold_val)], verbose=False)
                    
                    best_iteration_rmse = model.best_score # This is the RMSE from early stopping
                    fold_rmses.append(best_iteration_rmse)
                    
                avg_rmse = np.mean(fold_rmses)
                xgb_tuning_results.append({'params': current_params, 'avg_rmse': avg_rmse})

                if avg_rmse < best_xgb_rmse:
                    best_xgb_rmse = avg_rmse
                    best_xgb_params = current_params
                print(f"  Avg RMSE for this combo: {avg_rmse:.4f}")

print("\n--- XGBoost Tuning Complete ---")
print(f"Best XGBoost RMSE: {best_xgb_rmse:.4f}")
print(f"Best XGBoost Parameters: {best_xgb_params}")

# Train final XGBoost model with best parameters on the full training set
print("\n--- Training Final XGBoost Model ---")
# Pass X_train_xgb_tuned which already has label encoding applied
final_xgb_model = xgb.XGBRegressor(**best_xgb_params)
record_final_model_performance(final_xgb_model, "XGBoost (Tuned)", X_train_xgb_tuned, y_train_transformed, y_train_original,
                               X_test_xgb_tuned, y_test_transformed, y_test_original, best_xgb_params)
print("Final XGBoost Model Training Complete.")


# --- Model 3: Linear Regression (Naive/Baseline Solution) - No Tuning Here ---
print("\n--- Training Linear Regression (Naive) ---")
# Preprocessing for Linear Regression: StandardScaler for numerical, OneHotEncoder for categoricals
preprocessor_lr = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough'
)
lr_model = Pipeline(steps=[('preprocessor', preprocessor_lr),
                           ('regressor', LinearRegression(n_jobs=-1))])

# For Linear Regression, we don't tune (it has few hyperparameters that typically need tuning)
# Just record its performance on the full train/test split.
record_final_model_performance(lr_model, "LinearRegression", X_train, y_train_transformed, y_train_original,
                         X_test, y_test_transformed, y_test_original)
print("Linear Regression Training Complete.")

# --- Print Final Results Summary ---
print("\n--- Tuned Model Training & Performance Summary ---")
for model_name, model_data in results.items():
    print(f"\nModel: {model_name}")
    print(f"  Training Time: {model_data['train_time_sec']:.2f} seconds")
    print(f"  Prediction Time (Train Set): {model_data['predict_time_sec_train']:.4f} seconds")
    print(f"  Prediction Time (Test Set): {model_data['predict_time_sec_test']:.4f} seconds")

    print("  Train Metrics (Log Scale):")
    for metric, value in model_data['metrics_log_scale'].items():
        print(f"    {metric}: {value:.4f}")
    print("  Train Metrics (Original Scale):")
    for metric, value in model_data['metrics_original_scale'].items():
        print(f"    {metric}: {value:.4f}")

    print("  Test Metrics (Log Scale):")
    for metric, value in model_data['metrics_log_scale_test'].items():
        print(f"    {metric}: {value:.4f}")
    print("  Test Metrics (Original Scale):")
    for metric, value in model_data['metrics_original_scale_test'].items():
        print(f"    {metric}: {value:.4f}")
```
